#subplot
1. fig 도화지의 개념 그림의 밑바탕을 그리는 것
2. 실제 그림을 그리는 것은 axis ax를 통해서 그림을 그린다
- fig, ax=plt.sbuplots(nrows, ncols)

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(5,5))
#nrow: 가로줄 개수, ncols: 세로줄개수
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(5,5))
```

```python
fig, (ax1, ax2, ax3, ax4)=plt.subplots(nrows=4, ncols=1)
```

```python
fig, ax=plt.subplots(nrows=2, ncols=2)
ax[0][0].plot(mpg['cty'])
ax[0][1].hist(mpg['cty'], facecolor='purple')

#bar, scatter는 반드시 x축과 y축 데이터를 함께 입력해주어야 합니다.

ax[1][0].bar(mpg['cty'],mpg['hwy'], facecolor='red')
ax[1][1].scatter(mpg['cty'],mpg['hwy'], facecolor='green')
```

#gridspec

```python
import matplotlib.gridspec as gridspec
```

```python
mpg_features = ['manufacturer','model',
                'category','drv']

grid=gridspec.GridSpec(2,2)   #4개그래프

#plt.figure(figsize=(10,10))
#plt.subplots_adjust(wspace= 0.5, hspace= 0.4)


for t1, t2 in enumerate(mpg_features):
  sns.countplot(data=mpg,
                ax=plt.subplot(grid[t1]),
                x=t2
#               ,hue='drv'
#               ,palette = 'pastel'                
                )
#ax.set_title(f'{feature} Distribution')
```

```python
mpg_features = ['manufacturer','model',
                'category','drv']

grid=gridspec.GridSpec(2,2)

plt.figure(figsize=(10,10))
plt.subplots_adjust(wspace= 0.5, hspace= 0.4)


for t1, t2 in enumerate(mpg_features):
  sns.countplot(data=mpg,
                ax=plt.subplot(grid[t1]),
                x=t2
                ,hue='drv'
                ,palette = 'pastel'                
                )
#ax.set_title(f'{feature} Distribution')
```
