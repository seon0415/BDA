## na 값을 대체하는 방법!

### 다른 방법으로 누락값을 대체하는 법!
- fillna(method = '') 문법 공식
- ffill - 누락값이 나타나기 전의 값으로 누락값을 변경
- bfill - 누락값이 나타난 이후 첫 번째 값으로 앞쪽의 누락값을 변경

```python
ebola.fillna(method='ffill').iloc[0:10, 0:5]['Cases_Liberia']
```

```python
ebola.fillna(method='bfill').iloc[0:10, 0:5]['Cases_Liberia']
```

```python
ebola.iloc[0:10, 0:5]['Cases_Liberia'].mean()
```

```python
#보간법
ebola.interpolate().iloc[0:10, 0:5]['Cases_Liberia']
```

```python
#보간법
ebola.interpolate(method='quadratic').iloc[0:10, 0:5]['Cases_Liberia'].mean()
```

- 양쪽에 있는 값을 이용하여 중간값을 구한 다음 누락값을 처리한다.

## 누락값을 대체하는 법
- 보간법 
- interpolate(method='')
- 선형보간, spline 보간법 등등 

```python
ebola['Cases_Liberia'].plot()
```

```python
ebola.fillna(8049.2)['Cases_Liberia'].plot()
```

```python
ebola.interpolate()['Cases_Liberia'].plot()
```

- 선형보간법

```python
ebola.interpolate(method='linear')['Cases_Liberia'].plot()
```

- 스플라인 보간법
- 각 데이터의 포인트 구간별로 낮은 수준의 다항식 보간을 사용
- 직선으로 (선형이라는 건 직선으로 보간한다는 개념 미싱을 값을 양쪽 끝 값을 기준으로 직선을 긋는 )
- 직선으로 연결되는 것 보다느 좀 더 부드럽게 연결 된다.

```python
ebola.interpolate(method='quadratic')['Cases_Liberia'].plot()
```

```python
ebola.interpolate(method='dsfd')['Cases_Liberia'].plot()
```

method must be one of ['linear', 'time', 'index', 'values', 'nearest', 'zero', 'slinear', 'quadratic', 'cubic', 'barycentric', 'krogh', 'spline', 'polynomial', 'from_derivatives', 'piecewise_polynomial', 'pchip', 'akima', 'cubicspline']

- 시간에 대한 인덱스 기반도 있고, 다양한 보간법을 사용할 수 있다.

### melt 함수 
- melt는 모든 상황에 쓰는 게 아니라 melt를 써야 하는 상황에서 사용한다.
- melt 메서드 인자
- id_vars : 위치를 그대로 유지할 열의 이름을 지정
- value_vars : 행으로 위치를 변경할 열의 이름을 지정
- var_name :value_vars로 위치 변경할 열의 이름
- value_name : var_ name의 위치 변경할 열의 데이터를 저장한 열의 이름

```python
pew = pd.read_csv('pew.csv')
```

```python
pew
```

```python
pd.melt(pew, id_vars='religion') # id_vars  축을 뜻함
```

```python
pd.melt(pew, id_vars='religion', var_name ='income',value_name='count') 
```

### str 문법중 get 함수
- 값을 가지고 오는데 0,1,2 인덱스 순서로 값을 가지고 올 수 있다.

```python
case = vr_sp.str.get(0)
```

```python
case
```

```python
case_cr = vr_sp.str.get(1)
```

```python
ebola_sp['case']= case
ebola_sp['case_cr'] = case_cr
```

```python
ebola_sp
```
