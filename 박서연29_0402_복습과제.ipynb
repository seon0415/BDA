##시각화

```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt 
%matplotlib inline 
#매직커맨드 show함수 사용하지 않아도 바로 그래프가 나오는 것
```

###산점도

```python
plt.figure(figsize=(5,5))
plt.scatter([1,2,3],[4,5,6])
plt.show()
```

```python
#여러개
plt.scatter([1,2,3],[4,5,6])
plt.scatter([5,6,7],[10,5,6])
plt.scatter([2,4,8],[1,2,5])

plt.show()
```

```python
#그래프 축, 이름, 축범위 설정
plt.scatter([1,2,3],[4,5,6])

plt.title('mpg.cty') 
plt.xlabel('mpg.hwy')
plt.ylabel('mpg.cty') 

plt.xlim(0,10)
plt.ylim(3,8)
```

```python
#xticks
plt.scatter([1,2,3],[4,5,6])
plt.xticks([1,2,3],['A','B','C'])
```

```python
#범례
plt.figure(figsize=(5,5))
plt.scatter([1,2,3],[4,5,6],label='mpg.cty')
plt.scatter([6,7,8],[3,2,5],label='mpg.hwy')
plt.legend(loc='center')
```

###산점도 grid 종류

```python
import seaborn as sns
```

```python
sns.set() 
plt.scatter([1,2,3],[4,5,6]) 
```

```python
sns.set(style = 'dark')
plt.scatter([1,2,3],[4,5,6]) 
```

```python
sns.set(style = 'whitegrid') 
plt.scatter([1,2,3],[4,5,6]) 
```

```python
sns.set(style = 'white') 
plt.scatter([1,2,3],[4,5,6]) #그래프가 바뀌어진다.
```

###plot

```python
sns.set(style = 'whitegrid')
plt.plot([1,2,3],[4,5,6],label='mpg.cty')
plt.plot([6,7,8],[3,2,5],label='mpg.hwy')
plt.legend()
```

```python
#선 색깔
sns.set(style = 'whitegrid')
plt.plot([1,2,3],[4,5,6],label='mpg.cty',color='red')
plt.plot([6,7,8],[3,2,5],label='mpg.hwy',color='blue')
plt.plot([6,7,8],[3,1,4],label='mpg.car',color='yellow')
plt.legend(loc='center')

plt.xlabel('hi')
plt.ylabel('bye')
plt.xlim(1,10)
```

```python
#선스타일과 두께: dashed, dotted
sns.set(style = 'whitegrid')
plt.plot([1,2,3],[4,5,6],label='mpg.cty',color='red',linestyle='dashed', linewidth=10)
plt.plot([6,7,8],[3,2,5],label='mpg.hwy',color='blue',linestyle='dotted', linewidth=5)
plt.legend()
```

```python
#마커설정
sns.set(style = 'whitegrid')
plt.plot([1,2,3],[4,5,6],label='mpg.cty',color='red',linestyle='dashed', marker='s')
#점이 네모
plt.plot([6,7,8],[3,2,5],label='mpg.hwy',color='blue',linestyle='dotted', marker='*')
#점이 별
plt.plot([2,5,7],[3,2,5],label='mpg.hwy',color='navy',linestyle='dotted', marker='o', markersize=20)
#점이 원
plt.legend()
```
