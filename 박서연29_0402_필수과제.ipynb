## 필수과제 2
- 범례 값 만들어서 비교해 보기!
- left
- center
- right
- upper left , right, center
- lower left, right, center


```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt 
%matplotlib inline 

import seaborn as sns
```

```python
sns.set(style='white')
plt.scatter([1,2,4],[3,7,5], label='one')
plt.scatter([2,4,6],[1,3,4], label='two')

plt.legend(loc='lower left')
```

```python
sns.set(style='white')
plt.scatter([1,2,4],[3,7,5], label='three')
plt.scatter([2,4,6],[1,3,4], label='four')

plt.legend(loc='lower right')
```

```python
sns.set(style='dark')
plt.scatter([1,2,4],[3,7,5], label='one')
plt.scatter([2,4,6],[1,3,4], label='two')

plt.legend(loc='upper left')
```

```python
sns.set(style='dark')
plt.scatter([1,2,4],[3,7,5], label='five', color='red')
plt.scatter([2,4,6],[1,3,4], label='six', color='green')

plt.legend(loc='upper right')
```

```python
sns.set(style='ticks')
plt.scatter([1,2,4],[3,7,5], label='eight')
plt.scatter([2,4,6],[1,3,4], label='nine')

plt.legend(loc='right')
```

```python
sns.set(style='ticks')
plt.scatter([1,2,4],[3,7,5], label='eight', color='black')
plt.scatter([2,4,6],[1,3,4], label='nine')

plt.legend(loc='center')
```
