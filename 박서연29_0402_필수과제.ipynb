```python
import pandas as pd
mpg=pd.read_csv('mpg.csv')
mpg
```

```python
#append: 행을 추가하는 데이터프레임
fruit=pd.DataFrame({'a':['oragne','strawberry'],'b':['carrot','onion']})
fruit=fruit.append({'a':'grape','b':'tomato'},ignore_index=True)
print(fruit, "\n")

#get_values: get() 열의 값들 출력
print(fruit.get('a'))
print(fruit.get('b'))

#keys(): columns 출력
print("\n",fruit.keys())
print(list(fruit.keys()), "\n")

#values(): 값들 출력
a = {'name' : 'dobby', 'phone' : '010-1234-1234', 'address' : 'korea'}
print(a.values())
print(list(a.values()))

#to_frame: 데이터프레임으로 변환(리스트, 딕셔너리에서 변환가능)
df=pd.DataFrame(fruit)
df
```

```python
#describe: 요약통계량 출력
print(mpg.describe())

#drop_duplicates: 중복값 처리하기
print(mpg['manufacturer'].drop_duplicates())

#mean, max, min
print(mpg['cty'].mean(),'\n',
      mpg['cty'].max(),'\n',
      mpg['cty'].min())

#sort_values: 오름차순 혹은 내림차순 정렬
print(mpg['cty'].sort_values(ascending=False))
```

```python
#get_values
fruit=pd.DataFrame({'a':['oragne','strawberry'],'b':['carrot','onion']})

print(fruit.get('a'))
print(fruit.get('b'))
```

```python
mpg['manufacturer'].drop_duplicates()
```

```python
#equals
mpg[mpg['manufacturer']=='audi']

#isin: 원하는 원소가 있는지 확인해준다
mpg[mpg['manufacturer']=='hyundai'].isin(["sonata"])
```

```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt 
%matplotlib inline 

import seaborn as sns
```

```python
sns.set(style='white')
plt.scatter([1,2,4],[3,7,5], label='one')
plt.scatter([2,4,6],[1,3,4], label='two')

plt.legend(loc='lower left')
```

```python
sns.set(style='white')
plt.scatter([1,2,4],[3,7,5], label='three')
plt.scatter([2,4,6],[1,3,4], label='four')

plt.legend(loc='lower right')
```

```python
sns.set(style='dark')
plt.scatter([1,2,4],[3,7,5], label='one')
plt.scatter([2,4,6],[1,3,4], label='two')

plt.legend(loc='upper left')
```

```python
sns.set(style='dark')
plt.scatter([1,2,4],[3,7,5], label='five', color='red')
plt.scatter([2,4,6],[1,3,4], label='six', color='green')

plt.legend(loc='upper right')
```

```python
sns.set(style='ticks')
plt.scatter([1,2,4],[3,7,5], label='eight')
plt.scatter([2,4,6],[1,3,4], label='nine')

plt.legend(loc='right')
```

```python
sns.set(style='ticks')
plt.scatter([1,2,4],[3,7,5], label='eight', color='black')
plt.scatter([2,4,6],[1,3,4], label='nine')

plt.legend(loc='center')
```

## 필수과제 4 
- 위에 배운 내용을 그대로 적용하여 scatter를 다 그려주세요! ( 해당 인자값들 다 추가하셔야 합니다.)
- #scatter, #plot 동일하게 사용되는 것도 있고 다른 것도 있다. 에러가 날 것 -> 구글링해서 scatter 사용하는 인자값으로 바꿔야 한다.
- scatter 
- x와 y의 분포를 볼 때 정말 많이 사용한다.
- 데이터 시각화에서 기본적으로 변수관의 관계
- 비지도학습시 군집간의 관계, 차원축소 해서 해당 군집이 어떤 식으로 형성되어 있나 볼 때 
- 군집화하기전에 기본적으로 분포를 볼 때 사용하고 다양하게 사용된다!

```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)

a = np.random.normal(0,2,3000)
b = np.random.normal(0,2,3000)
```

```python
sns.set(style = 'dark')

plt.scatter(x,y, alpha=0.5)
plt.scatter(a,b, alpha=0.1)


plt.title('mpg.cty') 
plt.xlabel('mpg.hwy')
plt.ylabel('mpg.cty') 
plt.legend(loc='center')
```

```python
mpg=pd.read_csv('mpg.csv')
```

```python
sns.set(style='whitegrid')

plt.scatter(mpg.query('year==1999').cty,mpg.query('year==1999').hwy
            ,color='red',alpha=0.5, label='mpg.1999',marker='s', linewidth=5)
plt.scatter(mpg.query('year==2008').cty,mpg.query('year==2008').hwy,
            color='blue',alpha=0.5, label='mpg.2008', marker='*', linewidth=5)

plt.legend()
plt.colorbar()
plt.xlabel('mpg.cty')
plt.ylabel('mpg.hwy')
```
