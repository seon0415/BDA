아우디의 차량의 개수와, 아우디 차량의 hwy(고속도로 연비)는? 

```python
mpg[mpg['manufacturer']=='audi'].value_counts('manufacturer')
mpg[mpg['manufacturer']=='audi'].hwy
```

drv(구동방식)가 r에 해당하는 suv 차량의 개수

```python
mpg[mpg['drv']=='r'].query('category=="suv"').category.value_counts()
```

displ(배기량) 평균이 큰 manufacturer(자동차 제조회사) 1~3위

```python
mpg.groupby('manufacturer').displ.mean().sort_values(ascending=False).head(3)
```

displ (배기량)이 4이하인 자동차와 5 이상인 자동차의 hwy(고속 도로 연비) 평균

 "audi"와 "toyota"중에 어떤 manufacturer(자동차 제조 회사)의 cty(도시연비)

"chevrolet" , "ford", "honda" 자동차의 hwl(고속도로 연비) 평균

category가 "suv"인 자동차의 cty
