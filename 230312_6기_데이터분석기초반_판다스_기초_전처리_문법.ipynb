### 파생변수 만들기
- 파생변수는 기존의 변수들을 가지고 새로운 변수를 만들 수 있고, 기존의 컬럼 값을 가지고도 새로운 변수를 만들 수 있다.
- 파생변수는 새로운 변수지만 기존의 변수들의 관계를 가지고 만든다.

```python
## 평균을 만들자
# assign.(만들변수이름 = 식)
exam.assign(total_3 = (exam['math']+exam['english']+exam['science'])/3,
           sum_3 = exam['math']+exam['english']+exam['science'])

#그냥 셀에서 작동하고 나면 실제 원본 데이터는 반영이 되지 않는다./
#원본데이터를 바꾸고 싶으면 2가지 방법이 있다.

# exam_1 = exam.assign(total_3 = (exam['math']+exam['english']+exam['science'])/3,
    #       sum_3 = exam['math']+exam['english']+exam['science'])
# inplace=True
# exam.assign(total_3 = (exam['math']+exam['english']+exam['science'])/3,
    #       sum_3 = exam['math']+exam['english']+exam['science'], inplace=True)
#위에 2가지 방법으로 원본데이터를 바꿀 수 있다.
```

### groupby, agg 이용하기
- 엑셀에서 말하는 피벗 개념
- 피벗이라는 건 전체 데이터를 가지고 내가 원하는 컬럼만 요약통계를 볼 수 있다.
- 집단별로 요약통계를 구한다고 보면 됩니다.
- groupby라는 건 집단을 만드는 함수
- agg 다양한 통계치를 사용할 수 있는 함수

```python
exam.groupby('nclass').agg(mean_english = ('english','mean')) #파생변수이름, 사용할 컬럼(변수), 원하는 요약통계치
```

```python
exam.groupby('nclass').agg(mean_english = ('english','sum'))
```

```python
exam.groupby('nclass').agg(mean_english = ('english','median')) #max min 도 가능하다.
```

```python
#여러 개를 요약통계치 만들어 보자!
exam.groupby('nclass').agg(mean_math = ('math','mean'),
                         sum_math = ('math','sum'),
                         max_math = ('math','max'),
                         n = ('nclass','count')) #개수 카운팅  
#std 표준편차도 가능함
```

### merge
- 데이터 병합
- 깔끔한 데이터는 거의 대부분 없고 현업에서는 sql 데이터를 추출해야 하는데
- 데이터가 다른 테이블 있는 경우가 많고
- 두 개의 Df 추출해서 하나로 합쳐야 하는 경우가 있다.

- 데이터 분석 공모전에 나갔는데, 공공데이터를 같이 사용하고 싶어서
- 기존 공모전 데이터와 공공데이터를 붙여야 하는 경우?

- 데이터를 합친다는 것은 공통된 기준이 있어야 한다.

```python
a = pd.DataFrame({'name':['김철수','김영희','홍길동'],
                 '영어 점수':[10,20,30]})
b = pd.DataFrame({'name':['김철수','김영희','홍길길'],
                 '수학 점수':[10,20,40]})

```

```python
display(a)
display(b)
```

- merge
- concat 
- 이 두 가지는 1시간동안 병합에 대해서 따로 배울 예정

```python
#merge를 이용하면
#공통된 기준이 필요하다고 함
#공통된 기준은 name 의 이름이 겠죠? 김처루, 김영희 
total = pd.merge(a, b, how='inner') #name 공통된 키라고 이야기한다. pk 
```

```python
total
```

```python
#concat 
#덩어리 + 덩어리로 붙는 형태
pd.concat([a,b]) #axis 축 디폴트는 0값이죠 행기준
```

```python
pd.concat([a,b],axis=1) #1로 바꿈 
```

```python
display(a)
display(b)
```

- merge concat은 어떤 상황에서 사용하느냐에 따라 다르다.
- 두 개의 원리를 정확히 이해하고 있어야 대량의 데이터를 합칠 때 문제가 발생하지 않는다.
